```python
import numpy as np
import matplotlib.pyplot as plt

range_data = 420
t = np.arange(0, range_data, 1)

with open("tes_result2.txt", "r") as file:
    next(file)
    data = [next(file) for _ in range(range_data)]

setpoints = [float(line.split(',')[0]) for line in data]
yaw_values = [float(line.split(',')[1]) for line in data]

v = (83 - 15) / 8
P = [15, 20]

def mapping(t, z, P, v):
    x = []
    y = []
    k = len(t)
    for i in range(1, k):
        dt = t[i] - t[i-1]
        x_k = P[0] + v * np.cos(z[i]) * dt
        y_k = P[1] + v * np.sin(z[i]) * dt
        P = [x_k, y_k]
        x.append(x_k)
        y.append(y_k)

    return [x, y]

x, y = mapping(t, yaw_values, P, v)

# Plotting
plt.scatter(x, y, label='Robot Navigation', marker = '.')
plt.title('Robot Navigation')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```
